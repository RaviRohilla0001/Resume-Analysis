```python
import pdfplumber
import numpy as np
import spacy
nlp = spacy.load("en_core_web_sm")
import re
```

```python
def extract_text(file):
    text = ""
    with pdfplumber.open(file) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text
```

```python

def extract_basic_info(text):
    doc = nlp(text)
    name = ""
    for ent in doc.ents:
        if ent.label_ == "PERSON":
            name = ent.text
            break
    
    email = re.findall(r"[a-zA-Z0-9+_.-]+@[a-zA-Z0-9.-]+", text)
    
    return {
        "name": name,
        "email": email[0] if email else None
    }
```

```python
skills_db = [
    "python", "machine learning", "data science",
    "deep learning", "sql", "flask", "django"
]

def extract_skills(text):
    doc = nlp(text.lower())
    found_skills = set()
    
    for token in doc:
        if token.text in skills_db:
            found_skills.add(token.text)
    for skill in skills_db:
        if skill in text.lower():
            found_skills.add(skill)
    
    return list(found_skills)
```

```python
def calculate_similarity(resume_text, job_desc):
    doc1 = nlp(resume_text)
    doc2 = nlp(job_desc)
    return np.round(doc1.similarity(doc2) * 100,2)
```

```python
resume_text = extract_text(r"C:\Users\RAVI\Desktop\social media\R1.pdf")
job_desc = open("decription.txt").read()

info = extract_basic_info(resume_text)
skills = extract_skills(resume_text)
score = calculate_similarity(resume_text, job_desc)

print("Name:", info["name"])
print("Email:", info["email"])
print("Skills:", skills)
print("Match Score:", score)
```
